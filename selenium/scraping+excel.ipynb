```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import openpyxl

# Criar um novo arquivo Excel
wb = openpyxl.Workbook()
sheet = wb.active
sheet.title = "Links_Bolsonaro"

# Lista de URLs dos sites
sites = [
    "https://www.globo.com/",
    "https://www.uol.com.br/",
    "https://www.bbc.com/portuguese"
]

# Loop sobre os sites
for site in sites:
    # Inicializa o driver Chrome
    driver = webdriver.Chrome()

    try:
        # Abre o site
        driver.get(site)

        # Espera a página carregar e encontra todos os elementos de link no corpo
        links = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.XPATH, "//body//a")))

        # Itera sobre os links
        for link in links:
            try:
                # Tenta reencontrar o elemento
                link = WebDriverWait(driver, 2).until(EC.staleness_of(link))
                link = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//body//a")))
                
                # Se o texto do link contiver "Bolsonaro", escreve na planilha
                if "Bolsonaro" in link.text:
                    sheet.append([site, link.get_attribute('href')])
            except Exception as e:
                print(f"Erro ao reencontrar o elemento: {e}")

    except Exception as e:
        print(f"Ocorreu um erro para {site}: {e}")
    finally:
        # Fecha o driver
        driver.quit()

# Salva o arquivo Excel
wb.save("links_bolsonaro.xlsx")
```
